    A shallow copy creates a new object with the same values as the original object, but it doesn’t create copies of any objects that are referenced by the original object. Instead, it just copies the references to those objects.
    list.copy() -> it creates a shallow copy of list
    A deep copy, on the other hand, creates a new object with the same values as the original object and also creates copies of any objects that are referenced by the original object. This means that the new object will contain its own copies of any referenced objects, rather than just references to the same objects.
    copy.deepcopy(y) -> creates entire seprate copy

# 8. OPERATORS

    The Boolean data type represents two possible values: True and False. In Python, these values are written as 'True' and 'False', respectively.

The three Boolean operators are:

1. `and`:The `and` operator returns `True` if both operands are `True`, and `False` otherwise.

2. `or`: The `or` operator returns `True` if at least one of the operands is `True`, and `False` if both operands are `False`.

3. `not`: The `not` operator is a unary operator that returns the opposite of the operand's Boolean value. If the operand is `True`, `not` will return `False`, and if the operand is `False`, `not` will return `True`.



`and`:

| A     | B     | A and B |
|-------|-------|---------|
| True  | True  | True    |
| True  | False | False   |
| False | True  | False   |
| False | False | False   |

`or`:

| A    | B     | A or B |
|------|-------|--------|
| True | True  | True   |
| True | False | True   |
| False| True  | True   |
| False| False | False  |

`not`:

| A     | not A |
|-------|-------|
| True  | False |
| False | True  |

```python
#Write a Python program to input any alphabet and check whether it is vowel or consonant
def check_vowel_or_consonant(alphabet):
    vowels = ['a', 'e', 'i', 'o', 'u']
    if alphabet.lower() in vowels:
        return "Vowel"
    else:
        return "Consonant"

alphabet_input = input("Enter an alphabet: ")
if len(alphabet_input) == 1 and alphabet_input.isalpha():
    result = check_vowel_or_consonant(alphabet_input)
    print(f"The alphabet '{alphabet_input}' is a {result}.")
else:
    print("Invalid input. Please enter a single alphabet.")
```

```python
#3.Python Program to Find the Factorial of a Number
number = int(input("Enter a number: "))

factorial = 1

if number < 0:
    print("Factorial is not defined for negative numbers.")
elif number == 0:
    print("Factorial of 0 is 1.")
else:
    for i in range(1, number + 1):
        factorial *= i

    print("Factorial of", number, "is", factorial)
```

```python
#4.Python Program to Print the Fibonacci sequence
n = int(input("Enter the number of terms in the Fibonacci sequence: "))

# First two terms
a, b = 0, 1

if n <= 0:
    print("Invalid input. Please enter a positive integer.")
elif n == 1:
    print("Fibonacci sequence up to", n, "term:")
    print(a)
else:
    print("Fibonacci sequence up to", n, "terms:")
    print(a, b, end=" ")

    for _ in range(2, n):
        next_term = a + b
        print(next_term, end=" ")

        # Update values for the next iteration
        a, b = b, next_term
```

```python
# Guess game
import random

def guess_game():
    number = random.randint(1, 50)
    chances = 5

    print("Welcome to the Guess Game!")
    print("I've chosen a number between 1 and 50. You have 5 chances to guess it.")

    while chances > 0:
        guess = int(input("Enter your guess: "))

        if guess < 1 or guess > 50:
            print("Think within limits 1-50 only!")
            continue

        if guess < number:
            print("Hint: Think of a bigger number!")
        elif guess > number:
            print("Hint: Think of a smaller number!")
        else:
            print("Congratulations! You guessed the correct number!")
            return

        chances -= 1
        print("Chances remaining:", chances)

    print("Game over! You ran out of chances. The correct number was:", number)

guess_game()
```

```python
#BUBBLE SORT
def bubble_sort(arr):
    n = len(arr)

    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

numbers = [64, 34, 25, 12, 22, 11, 90]
bubble_sort(numbers)
print("Sorted array:", numbers)
```

```python
def selection_sort(arr):
    n = len(arr)

    for i in range(n):
        min_index = i
        for j in range(i+1, n):
            if arr[j] < arr[min_index]:
                min_index = j
    arr[i], arr[min_index] = arr[min_index], arr[i]

selection_sort(numbers)
print("Sorted array:", numbers)
    
```

```python
def insertion_sort(arr):
    n = len(arr)

    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key

insertion_sort(numbers)
print("Sorted array:", numbers)
```

```python
#3.Write a Python Program to remove punctuation from a given string (using loop)
def remove_punctuation(string):
    punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    result = ""
    for char in string:
        if char not in punctuations:
            result += char
    return result
string = "Hey there Priya! How you doing ?"
clean_string = remove_punctuation(string)

print("Original string:",string)
print("Cleaned string:", clean_string)
```

```python
#4.Write a Python Program to take a mxn matrix input from user and stored in nested list
def take_matrix_input(rows, columns):
    matrix = []

    # Iterate over each row
    for i in range(rows):
        row = []
        # Iterate over each column in the current row
        for j in range(columns):
            # Prompt the user to enter the element at position (i, j)
            element = int(input(f"Enter the element at position ({i+1}, {j+1}): "))
            row.append(element)
        # Append the current row to the matrix
        matrix.append(row)

    return matrix

num_rows = int(input("Enter the number of rows: "))
num_columns = int(input("Enter the number of columns: "))

matrix = take_matrix_input(num_rows, num_columns)

for row in matrix:
    print(row)
```

```python
#5.Write a Python Program to add, multiply and transpose two matrices given by user 
def get_matrix_dimensions():
    rows = int(input("Enter the number of rows: "))
    cols = int(input("Enter the number of columns: "))
    return rows, cols

def get_matrix_values(rows, cols):
    matrix = []
    print("Enter the matrix elements row by row:")
    for _ in range(rows):
        row = []
        for _ in range(cols):
            element = int(input())
            row.append(element)
        matrix.append(row)
    return matrix

def add_matrices(matrix1, matrix2):
    rows = len(matrix1)
    cols = len(matrix1[0])
    result = []
    for i in range(rows):
        row = []
        for j in range(cols):
            element = matrix1[i][j] + matrix2[i][j]
            row.append(element)
        result.append(row)
    return result

def multiply_matrices(matrix1, matrix2):
    rows1 = len(matrix1)
    cols1 = len(matrix1[0])
    rows2 = len(matrix2)
    cols2 = len(matrix2[0])

    if cols1 != rows2:
        raise ValueError("The number of columns in the first matrix must be equal to the number of rows in the second matrix.")

    result = []
    for i in range(rows1):
        row = []
        for j in range(cols2):
            element = 0
            for k in range(cols1):
                element += matrix1[i][k] * matrix2[k][j]
            row.append(element)
        result.append(row)
    return result

def transpose_matrix(matrix):
    rows = len(matrix)
    cols = len(matrix[0])
    result = []
    for j in range(cols):
        row = []
        for i in range(rows):
            element = matrix[i][j]
            row.append(element)
        result.append(row)
    return result

# Get dimensions of the matrices
print("Matrix 1:")
rows1, cols1 = get_matrix_dimensions()

print("Matrix 2:")
rows2, cols2 = get_matrix_dimensions()

# Get values of the matrices
print("Enter the values for Matrix 1:")
matrix1 = get_matrix_values(rows1, cols1)

print("Enter the values for Matrix 2:")
matrix2 = get_matrix_values(rows2, cols2)

# Perform operations
print("\nMatrix Addition:")
if rows1 == rows2 and cols1 == cols2:
    addition_result = add_matrices(matrix1, matrix2)
    for row in addition_result:
        print(row)

print("\nMatrix Multiplication:")
if cols1 == rows2:
    multiplication_result = multiply_matrices(matrix1, matrix2)
    for row in multiplication_result:
        print(row)

print("\nTranspose of Matrix 1:")
transpose_result1 = transpose_matrix(matrix1)
for row in transpose_result1:
    print(row)

print("\nTranspose of Matrix 2:")
transpose_result2 = transpose_matrix(matrix2)
for row in transpose_result2:
    print(row)
```

```python
#1.Program to find the sum of digits of a number until the sum is reduced to 1 digit
def sum_of_digits(num):
    while num > 9:
        num_str = str(num)
        digit_sum = 0
        for digit in num_str:
            digit_sum += int(digit)
        num = digit_sum
    return num

number = int(input("Enter a number: "))
result = sum_of_digits(number)
print("Sum of digits reduced to a single digit:", result)
```

```python
# Program to find LCM and HCF of two numbers
def calculate_lcm(num1,num2):
    if num1>num2:
        greater=num1
    else:
        greater=num2
    while(True):
        if((greater%num1==0) and (greater%num2==0)):
            lcm=greater
            break
        greater += 1

    return lcm

def calculate_hcf(num1,num2):
    while num2 != 0:
        num1, num2 = num2, num1 % num2

    return num1

# Get input from the user
number1 = int(input("Enter the first number: "))
number2 = int(input("Enter the second number: "))

# Calculate the LCM and HCF
lcm = calculate_lcm(number1, number2)
hcf = calculate_hcf(number1, number2)

# Display the results
print("LCM of", number1, "and", number2, "is:", lcm)
print("HCF of", number1, "and", number2, "is:", hcf)
    
    
```

```python
#3.Write a program to read roll number and marks of 10 students in 3 subjects. 
#The valid range for roll number is 1000-9999, if the roll number entered is not in this range, 
#the user should be asked to enter again. 
#Calculate total marks of only those students who get more than or equal to 40 marks in each subject. 
#Count the number of students whose total is more than 200. 
#Print the roll number of the students who gets the highest total.
def get_roll_number():
    while True:
        roll_number = int(input("Enter roll number (1000-9999): "))
        if 1000 <= roll_number <= 9999:
            return roll_number
        else:
            print("Invalid roll number. Please try again.")

def get_marks():
    while True:
        marks = int(input("Enter marks (0-100): "))
        if 0 <= marks <= 100:
            return marks
        else:
            print("Invalid marks. Please try again.")

# Initialize variables
total_marks = 0
count_total_more_than_200 = 0
highest_total = 0
highest_roll_number = 0

# Read roll number and marks of 10 students
for _ in range(10):
    print("Enter details for student", _+1)
    roll_number = get_roll_number()

    # Calculate total marks for each subject
    total_subjects = 0
    for _ in range(3):
        marks = get_marks()
        if marks >= 40:
            total_subjects += marks

    # Check conditions
    if total_subjects >= 120:
        total_marks += total_subjects
        if total_subjects > 200:
            count_total_more_than_200 += 1
        if total_subjects > highest_total:
            highest_total = total_subjects
            highest_roll_number = roll_number

# Print results
print("Total marks of students with >= 40 marks in each subject:", total_marks)
print("Number of students with total marks > 200:", count_total_more_than_200)
print("Roll number of student with highest total marks:", highest_roll_number)

```
